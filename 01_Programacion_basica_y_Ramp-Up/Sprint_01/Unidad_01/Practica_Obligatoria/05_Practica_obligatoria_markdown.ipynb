```python
df.rename(columns={ 
    "ID": "ID del Cliente", 
    "Age": "Edad del Cliente",
    "Experience":"Años Experiencia", 
    "Income":"Ingresos Anual",  
    "ZIP Code":"Codigo Postal", 
    "Family":"Miembros Familia",
    "CCAvg":"Promedio Gasto TDC", 
    "Education": "Nivel de Educación",
    "Mortgage": "Valor de Hipoteca",
    "Personal Loan":"Cliente ACCA",
    "Securities Account": "Cliente con Cuenta", 
    "CD Account": "Cliente con Deposito", 
    "Online":"Cliente Digital",
    "CreditCard": "Cliente con TDC"}, inplace=True)
```

```python
#PARA MEJOR COMPRENSIÓN LA COLUMNA INGRESOS ANUAL LA MULTIPLICAMOS POR 1000

df["Ingresos Anual"]= df["Ingresos Anual"]*1000
```

```python
#CREAMOS UNA NUEVA COLUMNA A PARTIR DE INGRESOS ANUAL LLAMADA INGRESOS MENSUAL 

df["Ingresos Mensual"] = (df["Ingresos Anual"]/12).astype(int)
```

```python
#Convertimos los datos de los datos de la columna Premedio gasto tdc 

df["Promedio Gasto TDC"]= (df["Promedio Gasto TDC"]*1000).astype(int)
```

```python
df["Valor de Hipoteca"]= (df["Valor de Hipoteca"]*1000).astype(int)
```

```python
mapeo_cliente_acca = {
    0: "No",
    1: "Si"
}

df["Cliente ACCA"]=df["Cliente ACCA"].map(mapeo_cliente_acca)
```

```python
mapeo_cliente_cuenta = {
    0: "No",
    1: "Si"
}

df["Cliente con Cuenta"]= df["Cliente con Cuenta"].map(mapeo_cliente_cuenta)
```

```python
mapeo_cliente_deposito = {
    0: "No",
    1: "Si"
}

df["Cliente con Deposito"]= df["Cliente con Deposito"].map(mapeo_cliente_deposito)
```

```python
mapeo_cliente_digital = {
    0: "No",
    1: "Si"
}

df["Cliente Digital"]=df["Cliente Digital"].map(mapeo_cliente_digital)
```

```python
mapeo_cliente_TDC = {
    0:"No",
    1 : "Si"
}

df["Cliente con TDC"]=df["Cliente con TDC"].map(mapeo_cliente_TDC)
```

```python
mapeo_nivel_educativo = {
    1: "No graduado",
    2: "Graduado",
    3: "Profesional"
}

df["Nivel de Educación"] = df["Nivel de Educación"].map(mapeo_nivel_educativo)

```

```python
def clasificar_edad(edad):
    if 18 <= edad <= 30:
        return "Cartera Joven"
    elif 31 <= edad <= 45:
        return "Adulto y Familias"
    elif 46 <= edad <= 60:
        return "Adultos Consolidados"
    elif edad > 60:
        return "Jubilados y Adultos Mayores"
    else:
        return "No Asignado"

df["Cartera Por Edad"] = df["Edad del Cliente"].apply(clasificar_edad)
```

```python
df
```

```python
#Clasificamos a los clientes segun su nivel de ingresos mensual 

def clasificar_cliente(ingresos):
    if ingresos <= 1000:
        return "Clientes Básicos"
    elif 1001 <= ingresos <= 4999:
        return "Banca Personal"
    elif 5000 <= ingresos <= 10000:
        return "Clientes Preferencial"
    elif ingresos > 10001:
        return "Clientes Oro"
    else:
        return "Sin Asignar"
    
df["Tipo de Cliente"]= df["Ingresos Mensual"].apply(clasificar_cliente)
```

```python
def plot_multiple_boxplots(df, columns, dim_matriz_visual = 2):
    num_cols = len(columns)
    num_rows = num_cols // dim_matriz_visual + num_cols % dim_matriz_visual
    fig, axes = plt.subplots(num_rows, dim_matriz_visual, figsize=(12, 6 * num_rows))
    axes = axes.flatten()

    for i, column in enumerate(columns):
        if df[column].dtype in ['int64', 'float64']:
            sns.boxplot(data=df, x=column, ax=axes[i])
            axes[i].set_title(column)

    # Ocultar ejes vacíos
    for j in range(i+1, num_rows * 2):
        axes[j].axis('off')

    plt.tight_layout()
    plt.show()
```

```python
def plot_histo_dens(df, columns, bins=None):
    num_cols = len(columns)
    num_rows = num_cols // 2 + num_cols % 2
    fig, axes = plt.subplots(num_rows, 2, figsize=(12, 6 * num_rows))
    axes = axes.flatten()

    for i, column in enumerate(columns):
        if df[column].dtype in ['int64', 'float64']:
            if bins:
                sns.histplot(df[column], kde=True, ax=axes[i], bins=bins)
            else:
                sns.histplot(df[column], kde=True, ax=axes[i])
            axes[i].set_title(f'Histograma y KDE de {column}')

    # Ocultar ejes vacíos
    for j in range(i + 1, num_rows * 2):
        axes[j].axis('off')

    plt.tight_layout()
    plt.show()
```

```python

def plot_combined_graphs(df, columns, whisker_width=1.5):
    num_cols = len(columns)
    if num_cols:
        
        fig, axes = plt.subplots(num_cols, 2, figsize=(12, 5 * num_cols))
        print(axes.shape)

        for i, column in enumerate(columns):
            if df[column].dtype in ['int64', 'float64']:
                # Histograma y KDE
                sns.histplot(df[column], kde=True, ax=axes[i,0] if num_cols > 1 else axes[0])
                if num_cols > 1:
                    axes[i,0].set_title(f'Histograma y KDE de {column}')
                else:
                    axes[0].set_title(f'Histograma y KDE de {column}')

                # Boxplot
                sns.boxplot(x=df[column], ax=axes[i,1] if num_cols > 1 else axes[1], whis=whisker_width)
                if num_cols > 1:
                    axes[i,1].set_title(f'Boxplot de {column}')
                else:
                    axes[1].set_title(f'Boxplot de {column}')

        plt.tight_layout()
        plt.show()
```

```python
plot_combined_graphs(df, ["Ingresos Mensual"])
```
